```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_breast_cancer, load_digits
from sklearn.model_selection import cross_val_score as cvs
from sklearn.naive_bayes import BernoulliNB, MultinomialNB, GaussianNB
```

```python
digits = pd.DataFrame(load_digits().data)
digits_y = pd.DataFrame(load_digits().target)
digits_forhead = digits
digits_forhead['target'] = digits_y
digits_forhead.head()
```

```python
bc = pd.DataFrame(load_breast_cancer().data)
bc_y = pd.DataFrame(load_breast_cancer().target)
bc_forhead = bc
bc_forhead['target'] = bc_y
bc_forhead.head()
```

```python
models = [BernoulliNB(), MultinomialNB(), GaussianNB()]
```

```python
for model in models:
    print(model, 'digits')
    print(cvs(model, digits, digits_y).mean())
    print(model, 'bc')
    print(cvs(model, bc, bc_y).mean())
    
```
