## –í–≤–µ–¥–µ–Ω–∏–µ
–õ–∏–Ω–µ–π–Ω–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è - –æ–¥–∏–Ω –∏–∑ –Ω–∞–∏–±–æ–ª–µ–µ —Ö–æ—Ä–æ—à–æ –∏–∑—É—á–µ–Ω–Ω—ã—Ö –º–µ—Ç–æ–¥–æ–≤ –º–∞—à–∏–Ω–Ω–æ–≥–æ –æ–±—É—á–µ–Ω–∏—è, –ø–æ–∑–≤–æ–ª—è—é—â–∏–π –ø—Ä–æ–≥–Ω–æ–∑–∏—Ä–æ–≤–∞—Ç—å –∑–Ω–∞—á–µ–Ω–∏—è –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞ –≤ –≤–∏–¥–µ –ª–∏–Ω–µ–π–Ω–æ–π –∫–æ–º–±–∏–Ω–∞—Ü–∏–∏ –ø—Ä–æ—á–∏—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ —Å –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏ - –≤–µ—Å–∞–º–∏ –º–æ–¥–µ–ª–∏. –û–ø—Ç–∏–º–∞–ª—å–Ω—ã–µ (–≤ —Å–º—ã—Å–ª–µ –º–∏–Ω–∏–º–∞–ª—å–Ω–æ—Å—Ç–∏ –Ω–µ–∫–æ—Ç–æ—Ä–æ–≥–æ —Ñ—É–Ω–∫—Ü–∏–æ–Ω–∞–ª–∞ –æ—à–∏–±–∫–∏) –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ –º–æ–∂–Ω–æ –Ω–∞–π—Ç–∏ –∞–Ω–∞–ª–∏—Ç–∏—á–µ—Å–∫–∏ —Å –ø–æ–º–æ—â—å—é –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —É—Ä–∞–≤–Ω–µ–Ω–∏—è –∏–ª–∏ —á–∏—Å–ª–µ–Ω–Ω–æ —Å –ø–æ–º–æ—â—å—é –º–µ—Ç–æ–¥–æ–≤ –æ–ø—Ç–∏–º–∏–∑–∞—Ü–∏–∏.  

–õ–∏–Ω–µ–π–Ω–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è –∏—Å–ø–æ–ª—å–∑—É–µ—Ç –ø—Ä–æ—Å—Ç–æ–π —Ñ—É–Ω–∫—Ü–∏–æ–Ω–∞–ª –∫–∞—á–µ—Å—Ç–≤–∞ - —Å—Ä–µ–¥–Ω–µ–∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω—É—é –æ—à–∏–±–∫—É. –ú—ã –±—É–¥–µ–º —Ä–∞–±–æ—Ç–∞—Ç—å —Å –≤—ã–±–æ—Ä–∫–æ–π, —Å–æ–¥–µ—Ä–∂–∞—â–µ–π 3 –ø—Ä–∏–∑–Ω–∞–∫–∞. –î–ª—è –Ω–∞—Å—Ç—Ä–æ–π–∫–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ (–≤–µ—Å–æ–≤) –º–æ–¥–µ–ª–∏ —Ä–µ—à–∞–µ—Ç—Å—è —Å–ª–µ–¥—É—é—â–∞—è –∑–∞–¥–∞—á–∞:
$$\Large \frac{1}{\ell}\sum_{i=1}^\ell{{((w_0 + w_1x_{i1} + w_2x_{i2} +  w_3x_{i3}) - y_i)}^2} \rightarrow \min_{w_0, w_1, w_2, w_3},$$
–≥–¥–µ $x_{i1}, x_{i2}, x_{i3}$ - –∑–Ω–∞—á–µ–Ω–∏—è –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ $i$-–≥–æ –æ–±—ä–µ–∫—Ç–∞, $y_i$ - –∑–Ω–∞—á–µ–Ω–∏–µ —Ü–µ–ª–µ–≤–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞ $i$-–≥–æ –æ–±—ä–µ–∫—Ç–∞, $\ell$ - —á–∏—Å–ª–æ –æ–±—ä–µ–∫—Ç–æ–≤ –≤ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–µ.

## –ì—Ä–∞–¥–∏–µ–Ω—Ç–Ω—ã–π —Å–ø—É—Å–∫
–ü–∞—Ä–∞–º–µ—Ç—Ä—ã $w_0, w_1, w_2, w_3$, –ø–æ –∫–æ—Ç–æ—Ä—ã–º –º–∏–Ω–∏–º–∏–∑–∏—Ä—É–µ—Ç—Å—è —Å—Ä–µ–¥–Ω–µ–∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–∞—è –æ—à–∏–±–∫–∞, –º–æ–∂–Ω–æ –Ω–∞—Ö–æ–¥–∏—Ç—å —á–∏—Å–ª–µ–Ω–Ω–æ —Å –ø–æ–º–æ—â—å—é –≥—Ä–∞–¥–∏–µ–Ω—Ç–Ω–æ–≥–æ —Å–ø—É—Å–∫–∞.
–ì—Ä–∞–¥–∏–µ–Ω—Ç–Ω—ã–π —à–∞–≥ –¥–ª—è –≤–µ—Å–æ–≤ –±—É–¥–µ—Ç –≤—ã–≥–ª—è–¥–µ—Ç—å —Å–ª–µ–¥—É—é—â–∏–º –æ–±—Ä–∞–∑–æ–º:
$$\Large w_0 \leftarrow w_0 - \frac{2\eta}{\ell} \sum_{i=1}^\ell{{((w_0 + w_1x_{i1} + w_2x_{i2} +  w_3x_{i3}) - y_i)}}$$
$$\Large w_j \leftarrow w_j - \frac{2\eta}{\ell} \sum_{i=1}^\ell{{x_{ij}((w_0 + w_1x_{i1} + w_2x_{i2} +  w_3x_{i3}) - y_i)}},\ j \in \{1,2,3\}$$
–ó–¥–µ—Å—å $\eta$ - –ø–∞—Ä–∞–º–µ—Ç—Ä, —à–∞–≥ –≥—Ä–∞–¥–∏–µ–Ω—Ç–Ω–æ–≥–æ —Å–ø—É—Å–∫–∞.

## –°—Ç–æ—Ö–∞—Å—Ç–∏—á–µ—Å–∫–∏–π –≥—Ä–∞–¥–∏–µ–Ω—Ç–Ω—ã–π —Å–ø—É—Å–∫
–ü—Ä–æ–±–ª–µ–º–∞ –≥—Ä–∞–¥–∏–µ–Ω—Ç–Ω–æ–≥–æ —Å–ø—É—Å–∫–∞, –æ–ø–∏—Å–∞–Ω–Ω–æ–≥–æ –≤—ã—à–µ, –≤ —Ç–æ–º, —á—Ç–æ –Ω–∞ –±–æ–ª—å—à–∏—Ö –≤—ã–±–æ—Ä–∫–∞—Ö —Å—á–∏—Ç–∞—Ç—å –Ω–∞ –∫–∞–∂–¥–æ–º —à–∞–≥–µ –≥—Ä–∞–¥–∏–µ–Ω—Ç –ø–æ –≤—Å–µ–º –∏–º–µ—é—â–∏–º—Å—è –¥–∞–Ω–Ω—ã–º –º–æ–∂–µ—Ç –±—ã—Ç—å –æ—á–µ–Ω—å –≤—ã—á–∏—Å–ª–∏—Ç–µ–ª—å–Ω–æ —Å–ª–æ–∂–Ω–æ. 
–í —Å—Ç–æ—Ö–∞—Å—Ç–∏—á–µ—Å–∫–æ–º –≤–∞—Ä–∏–∞–Ω—Ç–µ –≥—Ä–∞–¥–∏–µ–Ω—Ç–Ω–æ–≥–æ —Å–ø—É—Å–∫–∞ –ø–æ–ø—Ä–∞–≤–∫–∏ –¥–ª—è –≤–µ—Å–æ–≤ –≤—ã—á–∏—Å–ª—è—é—Ç—Å—è —Ç–æ–ª—å–∫–æ —Å —É—á–µ—Ç–æ–º –æ–¥–Ω–æ–≥–æ —Å–ª—É—á–∞–π–Ω–æ –≤–∑—è—Ç–æ–≥–æ –æ–±—ä–µ–∫—Ç–∞ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–∏:
$$\Large w_0 \leftarrow w_0 - \frac{2\eta}{\ell} {((w_0 + w_1x_{k1} + w_2x_{k2} +  w_3x_{k3}) - y_k)}$$
$$\Large w_j \leftarrow w_j - \frac{2\eta}{\ell} {x_{kj}((w_0 + w_1x_{k1} + w_2x_{k2} +  w_3x_{k3}) - y_k)},\ j \in \{1,2,3\},$$
–≥–¥–µ $k$ - —Å–ª—É—á–∞–π–Ω—ã–π –∏–Ω–¥–µ–∫—Å, $k \in \{1, \ldots, \ell\}$.

## –ù–æ—Ä–º–∞–ª—å–Ω–æ–µ —É—Ä–∞–≤–Ω–µ–Ω–∏–µ 
–ù–∞—Ö–æ–∂–¥–µ–Ω–∏–µ –≤–µ–∫—Ç–æ—Ä–∞ –æ–ø—Ç–∏–º–∞–ª—å–Ω—ã—Ö –≤–µ—Å–æ–≤ $w$ –º–æ–∂–µ—Ç –±—ã—Ç—å —Å–¥–µ–ª–∞–Ω–æ –∏ –∞–Ω–∞–ª–∏—Ç–∏—á–µ—Å–∫–∏.
–ú—ã —Ö–æ—Ç–∏–º –Ω–∞–π—Ç–∏ —Ç–∞–∫–æ–π –≤–µ–∫—Ç–æ—Ä –≤–µ—Å–æ–≤ $w$, —á—Ç–æ–±—ã –≤–µ–∫—Ç–æ—Ä $y$, –ø—Ä–∏–±–ª–∏–∂–∞—é—â–∏–π —Ü–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫, –ø–æ–ª—É—á–∞–ª—Å—è —É–º–Ω–æ–∂–µ–Ω–∏–µ–º –º–∞—Ç—Ä–∏—Ü—ã $X$ (—Å–æ—Å—Ç–æ—è—â–µ–π –∏–∑ –≤—Å–µ—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –æ–±—ä–µ–∫—Ç–æ–≤ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–∏, –∫—Ä–æ–º–µ —Ü–µ–ª–µ–≤–æ–≥–æ) –Ω–∞ –≤–µ–∫—Ç–æ—Ä –≤–µ—Å–æ–≤ $w$. –¢–æ –µ—Å—Ç—å, —á—Ç–æ–±—ã –≤—ã–ø–æ–ª–Ω—è–ª–æ—Å—å –º–∞—Ç—Ä–∏—á–Ω–æ–µ —É—Ä–∞–≤–Ω–µ–Ω–∏–µ:
$$\Large y = Xw$$
–î–æ–º–Ω–æ–∂–µ–Ω–∏–µ–º —Å–ª–µ–≤–∞ –Ω–∞ $X^T$ –ø–æ–ª—É—á–∞–µ–º:
$$\Large X^Ty = X^TXw$$
–≠—Ç–æ —Ö–æ—Ä–æ—à–æ, –ø–æ—Å–∫–æ–ª—å–∫—É —Ç–µ–ø–µ—Ä—å –º–∞—Ç—Ä–∏—Ü–∞ $X^TX$ - –∫–≤–∞–¥—Ä–∞—Ç–Ω–∞—è, –∏ –º–æ–∂–Ω–æ –Ω–∞–π—Ç–∏ —Ä–µ—à–µ–Ω–∏–µ (–≤–µ–∫—Ç–æ—Ä $w$) –≤ –≤–∏–¥–µ:
$$\Large w = {(X^TX)}^{-1}X^Ty$$
–ú–∞—Ç—Ä–∏—Ü–∞ ${(X^TX)}^{-1}X^T$ - [*–ø—Å–µ–≤–¥–æ–æ–±—Ä–∞—Ç–Ω–∞—è*](https://ru.wikipedia.org/wiki/–ü—Å–µ–≤–¥–æ–æ–±—Ä–∞—Ç–Ω–∞—è_–º–∞—Ç—Ä–∏—Ü–∞) –¥–ª—è –º–∞—Ç—Ä–∏—Ü—ã $X$. –í NumPy —Ç–∞–∫—É—é –º–∞—Ç—Ä–∏—Ü—É –º–æ–∂–Ω–æ –≤—ã—á–∏—Å–ª–∏—Ç—å —Å –ø–æ–º–æ—â—å—é —Ñ—É–Ω–∫—Ü–∏–∏ [numpy.linalg.pinv](http://docs.scipy.org/doc/numpy-1.10.0/reference/generated/numpy.linalg.pinv.html).

–û–¥–Ω–∞–∫–æ, –Ω–∞—Ö–æ–∂–¥–µ–Ω–∏–µ –ø—Å–µ–≤–¥–æ–æ–±—Ä–∞—Ç–Ω–æ–π –º–∞—Ç—Ä–∏—Ü—ã - –æ–ø–µ—Ä–∞—Ü–∏—è –≤—ã—á–∏—Å–ª–∏—Ç–µ–ª—å–Ω–æ —Å–ª–æ–∂–Ω–∞—è –∏ –Ω–µ—Å—Ç–∞–±–∏–ª—å–Ω–∞—è –≤ —Å–ª—É—á–∞–µ –º–∞–ª–æ–≥–æ –æ–ø—Ä–µ–¥–µ–ª–∏—Ç–µ–ª—è –º–∞—Ç—Ä–∏—Ü—ã $X$ (–ø—Ä–æ–±–ª–µ–º–∞ –º—É–ª—å—Ç–∏–∫–æ–ª–ª–∏–Ω–µ–∞—Ä–Ω–æ—Å—Ç–∏). 
–ù–∞ –ø—Ä–∞–∫—Ç–∏–∫–µ –ª—É—á—à–µ –Ω–∞—Ö–æ–¥–∏—Ç—å –≤–µ–∫—Ç–æ—Ä –≤–µ—Å–æ–≤ $w$ —Ä–µ—à–µ–Ω–∏–µ–º –º–∞—Ç—Ä–∏—á–Ω–æ–≥–æ —É—Ä–∞–≤–Ω–µ–Ω–∏—è

–ù–æ –≤—Å–µ –∂–µ –Ω–∞ –ø—Ä–∞–∫—Ç–∏–∫–µ –¥–ª—è –±–æ–ª—å—à–∏—Ö –º–∞—Ç—Ä–∏—Ü $X$ –±—ã—Å—Ç—Ä–µ–µ —Ä–∞–±–æ—Ç–∞–µ—Ç –≥—Ä–∞–¥–∏–µ–Ω—Ç–Ω—ã–π —Å–ø—É—Å–∫, –æ—Å–æ–±–µ–Ω–Ω–æ –µ–≥–æ —Å—Ç–æ—Ö–∞—Å—Ç–∏—á–µ—Å–∫–∞—è –≤–µ—Ä—Å–∏—è.

**–î–æ–±–∞–≤—å—Ç–µ –∫ –º–∞—Ç—Ä–∏—Ü–µ *X* —Å—Ç–æ–ª–±–µ—Ü –∏–∑ –µ–¥–∏–Ω–∏—Ü, –∏—Å–ø–æ–ª—å–∑—É—è –º–µ—Ç–æ–¥—ã *hstack*, *ones* –∏ *reshape* –±–∏–±–ª–∏–æ—Ç–µ–∫–∏ NumPy. –í–µ–∫—Ç–æ—Ä –∏–∑ –µ–¥–∏–Ω–∏—Ü –Ω—É–∂–µ–Ω –¥–ª—è —Ç–æ–≥–æ, —á—Ç–æ–±—ã –Ω–µ –æ–±—Ä–∞–±–∞—Ç—ã–≤–∞—Ç—å –æ—Ç–¥–µ–ª—å–Ω–æ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç $w_0$ –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏.**

```python
X2 = np.hstack((np.ones(200).reshape(200,1), X1))
```

**2. –†–µ–∞–ª–∏–∑—É–π—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é *mserror* - —Å—Ä–µ–¥–Ω–µ–∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω—É—é –æ—à–∏–±–∫—É –ø—Ä–æ–≥–Ω–æ–∑–∞. –û–Ω–∞ –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –¥–≤–∞ –∞—Ä–≥—É–º–µ–Ω—Ç–∞ - –æ–±—ä–µ–∫—Ç—ã Series *y* (–∑–Ω–∞—á–µ–Ω–∏—è —Ü–µ–ª–µ–≤–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞) –∏ *y\_pred* (–ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è). –ù–µ –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ –≤ —ç—Ç–æ–π —Ñ—É–Ω–∫—Ü–∏–∏ —Ü–∏–∫–ª—ã - —Ç–æ–≥–¥–∞ –æ–Ω–∞ –±—É–¥–µ—Ç –≤—ã—á–∏—Å–ª–∏—Ç–µ–ª—å–Ω–æ –Ω–µ—ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ–π.**

```python
def mserror(y, y_pred):
    return (np.sum((y_pred - y)**2)) / y.shape[0]
```

**–ö–∞–∫–æ–≤–∞ —Å—Ä–µ–¥–Ω–µ–∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–∞—è –æ—à–∏–±–∫–∞ –ø—Ä–æ–≥–Ω–æ–∑–∞ –∑–Ω–∞—á–µ–Ω–∏–π Sales, –µ—Å–ª–∏ –≤—Å–µ–≥–¥–∞ –ø—Ä–µ–¥—Å–∫–∞–∑—ã–≤–∞—Ç—å –º–µ–¥–∏–∞–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ Sales –ø–æ –∏—Å—Ö–æ–¥–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ? –ü–æ–ª—É—á–µ–Ω–Ω—ã–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç, –æ–∫—Ä—É–≥–ª–µ–Ω–Ω—ã–π –¥–æ 3 –∑–Ω–∞–∫–æ–≤ –ø–æ—Å–ª–µ –∑–∞–ø—è—Ç–æ–π, —è–≤–ª—è–µ—Ç—Å—è –æ—Ç–≤–µ—Ç–æ–º –Ω–∞ *'1 –∑–∞–¥–∞–Ω–∏–µ'.***

```python
answer1 = mserror(y, np.median(y))
print(round(answer1, 3))
```

**3. –†–µ–∞–ª–∏–∑—É–π—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é *normal_equation*, –∫–æ—Ç–æ—Ä–∞—è –ø–æ –∑–∞–¥–∞–Ω–Ω—ã–º –º–∞—Ç—Ä–∏—Ü–∞–º (–º–∞—Å—Å–∏–≤–∞–º NumPy) *X* –∏ *y* –≤—ã—á–∏—Å–ª—è–µ—Ç –≤–µ–∫—Ç–æ—Ä –≤–µ—Å–æ–≤ $w$ —Å–æ–≥–ª–∞—Å–Ω–æ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–º—É —É—Ä–∞–≤–Ω–µ–Ω–∏—é –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏.**

```python
def normal_equation(X, y):
    #ùë§=(ùëãùëáùëã)‚àí1 ùëãùëáùë¶
    Xt = np.transpose(X2)
    XtX = Xt.dot(X2)
    Xty = Xt.dot(y)
    XtX1 = np.linalg.inv(XtX)
    return XtX1.dot(Xty)
```

```python
norm_eq_weights = normal_equation(X2, y)
print(norm_eq_weights)
```

**–ö–∞–∫–∏–µ –ø—Ä–æ–¥–∞–∂–∏ –ø—Ä–µ–¥—Å–∫–∞–∑—ã–≤–∞—é—Ç—Å—è –ª–∏–Ω–µ–π–Ω–æ–π –º–æ–¥–µ–ª—å—é —Å –≤–µ—Å–∞–º–∏, –Ω–∞–π–¥–µ–Ω–Ω—ã–º–∏ —Å –ø–æ–º–æ—â—å—é –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —É—Ä–∞–≤–Ω–µ–Ω–∏—è, –≤ —Å–ª—É—á–∞–µ —Å—Ä–µ–¥–Ω–∏—Ö –∏–Ω–≤–µ—Å—Ç–∏—Ü–∏–π –≤ —Ä–µ–∫–ª–∞–º—É –ø–æ –¢–í, —Ä–∞–¥–∏–æ –∏ –≤ –≥–∞–∑–µ—Ç–∞—Ö? (—Ç–æ –µ—Å—Ç—å –ø—Ä–∏ –Ω—É–ª–µ–≤—ã—Ö –∑–Ω–∞—á–µ–Ω–∏—è—Ö –º–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ TV, Radio –∏ Newspaper). –ü–æ–ª—É—á–µ–Ω–Ω—ã–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç, –æ–∫—Ä—É–≥–ª–µ–Ω–Ω—ã–π –¥–æ 3 –∑–Ω–∞–∫–æ–≤ –ø–æ—Å–ª–µ –∑–∞–ø—è—Ç–æ–π, —è–≤–ª—è–µ—Ç—Å—è –æ—Ç–≤–µ—Ç–æ–º –Ω–∞ *'2 –∑–∞–¥–∞–Ω–∏–µ'*.**

**4. –ù–∞–ø–∏—à–∏—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é *linear_prediction*, –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –Ω–∞ –≤—Ö–æ–¥ –º–∞—Ç—Ä–∏—Ü—É *X* –∏ –≤–µ–∫—Ç–æ—Ä –≤–µ—Å–æ–≤ –ª–∏–Ω–µ–π–Ω–æ–π –º–æ–¥–µ–ª–∏ *w*, –∞ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –≤–µ–∫—Ç–æ—Ä –ø—Ä–æ–≥–Ω–æ–∑–æ–≤ –≤ –≤–∏–¥–µ –ª–∏–Ω–µ–π–Ω–æ–π –∫–æ–º–±–∏–Ω–∞—Ü–∏–∏ —Å—Ç–æ–ª–±—Ü–æ–≤ –º–∞—Ç—Ä–∏—Ü—ã *X* —Å –≤–µ—Å–∞–º–∏ *w*.**

```python
def linear_prediction(X, w):
    return X.dot(w)
```

**–ö–∞–∫–æ–≤–∞ —Å—Ä–µ–¥–Ω–µ–∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–∞—è –æ—à–∏–±–∫–∞ –ø—Ä–æ–≥–Ω–æ–∑–∞ –∑–Ω–∞—á–µ–Ω–∏–π Sales –≤ –≤–∏–¥–µ –ª–∏–Ω–µ–π–Ω–æ–π –º–æ–¥–µ–ª–∏ —Å –≤–µ—Å–∞–º–∏, –Ω–∞–π–¥–µ–Ω–Ω—ã–º–∏ —Å –ø–æ–º–æ—â—å—é –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —É—Ä–∞–≤–Ω–µ–Ω–∏—è?
–ü–æ–ª—É—á–µ–Ω–Ω—ã–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç, –æ–∫—Ä—É–≥–ª–µ–Ω–Ω—ã–π –¥–æ 3 –∑–Ω–∞–∫–æ–≤ –ø–æ—Å–ª–µ –∑–∞–ø—è—Ç–æ–π, —è–≤–ª—è–µ—Ç—Å—è –æ—Ç–≤–µ—Ç–æ–º –Ω–∞ *'3 –∑–∞–¥–∞–Ω–∏–µ'***

```python
answer3 = mserror(y, linear_prediction(X2, norm_eq_weights))
print(round(answer3, 3))
```

**5. –ù–∞–ø–∏—à–∏—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é *stochastic_gradient_step*, —Ä–µ–∞–ª–∏–∑—É—é—â—É—é —à–∞–≥ —Å—Ç–æ—Ö–∞—Å—Ç–∏—á–µ—Å–∫–æ–≥–æ –≥—Ä–∞–¥–∏–µ–Ω—Ç–Ω–æ–≥–æ —Å–ø—É—Å–∫–∞ –¥–ª—è –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏. –§—É–Ω–∫—Ü–∏—è –¥–æ–ª–∂–Ω–∞ –ø—Ä–∏–Ω–∏–º–∞—Ç—å –º–∞—Ç—Ä–∏—Ü—É *X*, –≤–µ–∫—Ç–æ—Ä–∞ *y* –∏ *w*, —á–∏—Å–ª–æ *train_ind* - –∏–Ω–¥–µ–∫—Å –æ–±—ä–µ–∫—Ç–∞ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–∏ (—Å—Ç—Ä–æ–∫–∏ –º–∞—Ç—Ä–∏—Ü—ã *X*), –ø–æ –∫–æ—Ç–æ—Ä–æ–º—É —Å—á–∏—Ç–∞–µ—Ç—Å—è –∏–∑–º–µ–Ω–µ–Ω–∏–µ –≤–µ—Å–æ–≤, –∞ —Ç–∞–∫–∂–µ —á–∏—Å–ª–æ *$\eta$* (eta) - —à–∞–≥ –≥—Ä–∞–¥–∏–µ–Ω—Ç–Ω–æ–≥–æ —Å–ø—É—Å–∫–∞ (–ø–æ —É–º–æ–ª—á–∞–Ω–∏—é *eta*=0.01). –†–µ–∑—É–ª—å—Ç–∞—Ç–æ–º –±—É–¥–µ—Ç –≤–µ–∫—Ç–æ—Ä –æ–±–Ω–æ–≤–ª–µ–Ω–Ω—ã—Ö –≤–µ—Å–æ–≤. –ù–∞—à–∞ —Ä–µ–∞–ª–∏–∑–∞—Ü–∏—è —Ñ—É–Ω–∫—Ü–∏–∏ –±—É–¥–µ—Ç —è–≤–Ω–æ –Ω–∞–ø–∏—Å–∞–Ω–∞ –¥–ª—è –¥–∞–Ω–Ω—ã—Ö —Å 3 –ø—Ä–∏–∑–Ω–∞–∫–∞–º–∏, –Ω–æ –Ω–µ—Å–ª–æ–∂–Ω–æ –º–æ–¥–∏—Ñ–∏—Ü–∏—Ä–æ–≤–∞—Ç—å –¥–ª—è –ª—é–±–æ–≥–æ —á–∏—Å–ª–∞ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤, –º–æ–∂–µ—Ç–µ —ç—Ç–æ —Å–¥–µ–ª–∞—Ç—å.**

## –°—Ç–æ—Ö–∞—Å—Ç–∏—á–µ—Å–∫–∏–π –≥—Ä–∞–¥–∏–µ–Ω—Ç–Ω—ã–π —Å–ø—É—Å–∫
–ü—Ä–æ–±–ª–µ–º–∞ –≥—Ä–∞–¥–∏–µ–Ω—Ç–Ω–æ–≥–æ —Å–ø—É—Å–∫–∞, –æ–ø–∏—Å–∞–Ω–Ω–æ–≥–æ –≤—ã—à–µ, –≤ —Ç–æ–º, —á—Ç–æ –Ω–∞ –±–æ–ª—å—à–∏—Ö –≤—ã–±–æ—Ä–∫–∞—Ö —Å—á–∏—Ç–∞—Ç—å –Ω–∞ –∫–∞–∂–¥–æ–º —à–∞–≥–µ –≥—Ä–∞–¥–∏–µ–Ω—Ç –ø–æ –≤—Å–µ–º –∏–º–µ—é—â–∏–º—Å—è –¥–∞–Ω–Ω—ã–º –º–æ–∂–µ—Ç –±—ã—Ç—å –æ—á–µ–Ω—å –≤—ã—á–∏—Å–ª–∏—Ç–µ–ª—å–Ω–æ —Å–ª–æ–∂–Ω–æ. 
–í —Å—Ç–æ—Ö–∞—Å—Ç–∏—á–µ—Å–∫–æ–º –≤–∞—Ä–∏–∞–Ω—Ç–µ –≥—Ä–∞–¥–∏–µ–Ω—Ç–Ω–æ–≥–æ —Å–ø—É—Å–∫–∞ –ø–æ–ø—Ä–∞–≤–∫–∏ –¥–ª—è –≤–µ—Å–æ–≤ –≤—ã—á–∏—Å–ª—è—é—Ç—Å—è —Ç–æ–ª—å–∫–æ —Å —É—á–µ—Ç–æ–º –æ–¥–Ω–æ–≥–æ —Å–ª—É—á–∞–π–Ω–æ –≤–∑—è—Ç–æ–≥–æ –æ–±—ä–µ–∫—Ç–∞ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–∏:
$$\Large w_0 \leftarrow w_0 - \frac{2\eta}{\ell} {((w_0 + w_1x_{k1} + w_2x_{k2} +  w_3x_{k3}) - y_k)}$$
$$\Large w_j \leftarrow w_j - \frac{2\eta}{\ell} {x_{kj}((w_0 + w_1x_{k1} + w_2x_{k2} +  w_3x_{k3}) - y_k)},\ j \in \{1,2,3\},$$
–≥–¥–µ $k$ - —Å–ª—É—á–∞–π–Ω—ã–π –∏–Ω–¥–µ–∫—Å, $k \in \{1, \ldots, \ell\}$.

```python
def stochastic_gradient_step(X, y, w, train_ind, eta=0.01):
    
    grad0 = w[0] - (1 / 200) * ((w[0] + 
                                       w[1] * X[train_ind][1] + 
                                       w[2] * X[train_ind][2] + 
                                       w[3] * X[train_ind][3]) - y[train_ind])
    
    grad1 = w[1] - (1 / 200) * X[train_ind][1] * ((w[0] + 
                                       w[1] * X[train_ind][1] + 
                                       w[2] * X[train_ind][2] + 
                                       w[3] * X[train_ind][3]) - y[train_ind])
    
    grad2 = w[2] - (1 / 200) * X[train_ind][2] * ((w[0] + 
                                       w[1] * X[train_ind][1] + 
                                       w[2] * X[train_ind][2] + 
                                       w[3] * X[train_ind][3]) - y[train_ind])
    
    grad3 = w[3] - (1 / 200) * X[train_ind][3] * ((w[0] + 
                                       w[1] * X[train_ind][1] + 
                                       w[2] * X[train_ind][2] + 
                                       w[3] * X[train_ind][3]) - y[train_ind])
    
    return  np.array([grad0, grad1, grad2, grad3])
```

**6. –ù–∞–ø–∏—à–∏—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é *stochastic_gradient_descent*, —Ä–µ–∞–ª–∏–∑—É—é—â—É—é —Å—Ç–æ—Ö–∞—Å—Ç–∏—á–µ—Å–∫–∏–π –≥—Ä–∞–¥–∏–µ–Ω—Ç–Ω—ã–π —Å–ø—É—Å–∫ –¥–ª—è –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏. –§—É–Ω–∫—Ü–∏—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –Ω–∞ –≤—Ö–æ–¥ —Å–ª–µ–¥—É—é—â–∏–µ –∞—Ä–≥—É–º–µ–Ω—Ç—ã:**
- X - –º–∞—Ç—Ä–∏—Ü–∞, —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∞—è –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–µ
- y - –≤–µ–∫—Ç–æ—Ä –∑–Ω–∞—á–µ–Ω–∏–π —Ü–µ–ª–µ–≤–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞
- w_init - –≤–µ–∫—Ç–æ—Ä –Ω–∞—á–∞–ª—å–Ω—ã—Ö –≤–µ—Å–æ–≤ –º–æ–¥–µ–ª–∏
- eta - —à–∞–≥ –≥—Ä–∞–¥–∏–µ–Ω—Ç–Ω–æ–≥–æ —Å–ø—É—Å–∫–∞ (–ø–æ —É–º–æ–ª—á–∞–Ω–∏—é 0.01)
- max_iter - –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ –∏—Ç–µ—Ä–∞—Ü–∏–π –≥—Ä–∞–¥–∏–µ–Ω—Ç–Ω–æ–≥–æ —Å–ø—É—Å–∫–∞ (–ø–æ —É–º–æ–ª—á–∞–Ω–∏—é 10000)
- max_weight_dist - –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –µ–≤–∫–ª–∏–¥–æ–≤–æ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –º–µ–∂–¥—É –≤–µ–∫—Ç–æ—Ä–∞–º–∏ –≤–µ—Å–æ–≤ –Ω–∞ —Å–æ—Å–µ–¥–Ω–∏—Ö –∏—Ç–µ—Ä–∞—Ü–∏—è—Ö –≥—Ä–∞–¥–∏–µ–Ω—Ç–Ω–æ–≥–æ —Å–ø—É—Å–∫–∞,
–ø—Ä–∏ –∫–æ—Ç–æ—Ä–æ–º –∞–ª–≥–æ—Ä–∏—Ç–º –ø—Ä–µ–∫—Ä–∞—â–∞–µ—Ç —Ä–∞–±–æ—Ç—É (–ø–æ —É–º–æ–ª—á–∞–Ω–∏—é 1e-8)
- seed - —á–∏—Å–ª–æ, –∏—Å–ø–æ–ª—å–∑—É–µ–º–æ–µ –¥–ª—è –≤–æ—Å–ø—Ä–æ–∏–∑–≤–æ–¥–∏–º–æ—Å—Ç–∏ —Å–≥–µ–Ω–µ—Ä–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –ø—Å–µ–≤–¥–æ—Å–ª—É—á–∞–π–Ω—ã—Ö —á–∏—Å–µ–ª (–ø–æ —É–º–æ–ª—á–∞–Ω–∏—é 42)

**–ù–∞ –∫–∞–∂–¥–æ–π –∏—Ç–µ—Ä–∞—Ü–∏–∏ –≤ –≤–µ–∫—Ç–æ—Ä (—Å–ø–∏—Å–æ–∫) –¥–æ–ª–∂–Ω–æ –∑–∞–ø–∏—Å—ã–≤–∞—Ç—å—Å—è —Ç–µ–∫—É—â–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —Å—Ä–µ–¥–Ω–µ–∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ–π –æ—à–∏–±–∫–∏. –§—É–Ω–∫—Ü–∏—è –¥–æ–ª–∂–Ω–∞ –≤–æ–∑–≤—Ä–∞—â–∞—Ç—å –≤–µ–∫—Ç–æ—Ä –≤–µ—Å–æ–≤ $w$, –∞ —Ç–∞–∫–∂–µ –≤–µ–∫—Ç–æ—Ä (—Å–ø–∏—Å–æ–∫) –æ—à–∏–±–æ–∫.**

```python
def stochastic_gradient_descent(X, y, w_init, eta=0.01, max_iter=1e4,
                                min_weight_dist=1e-8, seed=42, verbose=False):
    # –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –º–µ–∂–¥—É –≤–µ–∫—Ç–æ—Ä–∞–º–∏ –≤–µ—Å–æ–≤ –Ω–∞ —Å–æ—Å–µ–¥–Ω–∏—Ö
    # –∏—Ç–µ—Ä–∞—Ü–∏—è—Ö –±–æ–ª—å—à–∏–º —á–∏—Å–ª–æ–º. 
    weight_dist = np.inf
    # –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º –≤–µ–∫—Ç–æ—Ä –≤–µ—Å–æ–≤
    w = w_init
    # –°—é–¥–∞ –±—É–¥–µ–º –∑–∞–ø–∏—Å—ã–≤–∞—Ç—å –æ—à–∏–±–∫–∏ –Ω–∞ –∫–∞–∂–¥–æ–π –∏—Ç–µ—Ä–∞—Ü–∏–∏
    errors = []
    # –°—á–µ—Ç—á–∏–∫ –∏—Ç–µ—Ä–∞—Ü–∏–π
    iter_num = 0
    # –ë—É–¥–µ–º –ø–æ—Ä–æ–∂–¥–∞—Ç—å –ø—Å–µ–≤–¥–æ—Å–ª—É—á–∞–π–Ω—ã–µ —á–∏—Å–ª–∞ 
    # (–Ω–æ–º–µ—Ä –æ–±—ä–µ–∫—Ç–∞, –∫–æ—Ç–æ—Ä—ã–π –±—É–¥–µ—Ç –º–µ–Ω—è—Ç—å –≤–µ—Å–∞), –∞ –¥–ª—è –≤–æ—Å–ø—Ä–æ–∏–∑–≤–æ–¥–∏–º–æ—Å—Ç–∏
    # —ç—Ç–æ–π –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ –ø—Å–µ–≤–¥–æ—Å–ª—É—á–∞–π–Ω—ã—Ö —á–∏—Å–µ–ª –∏—Å–ø–æ–ª—å–∑—É–µ–º seed.
    np.random.seed(seed)
        
    # –û—Å–Ω–æ–≤–Ω–æ–π —Ü–∏–∫–ª
    while weight_dist > min_weight_dist and iter_num < max_iter:
        # –ø–æ—Ä–æ–∂–¥–∞–µ–º –ø—Å–µ–≤–¥–æ—Å–ª—É—á–∞–π–Ω—ã–π 
        # –∏–Ω–¥–µ–∫—Å –æ–±—ä–µ–∫—Ç–∞ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–∏
        random_ind = np.random.randint(X.shape[0])
        
        errors.append(mserror(y, linear_prediction(X, w)))
        w = stochastic_gradient_step(X, y, w, random_ind, eta) 
        iter_num += 1
        
    return w, errors
```

 **–ó–∞–ø—É—Å—Ç–∏—Ç–µ $10^5$ –∏—Ç–µ—Ä–∞—Ü–∏–π —Å—Ç–æ—Ö–∞—Å—Ç–∏—á–µ—Å–∫–æ–≥–æ –≥—Ä–∞–¥–∏–µ–Ω—Ç–Ω–æ–≥–æ —Å–ø—É—Å–∫–∞. –£–∫–∞–∂–∏—Ç–µ –≤–µ–∫—Ç–æ—Ä –Ω–∞—á–∞–ª—å–Ω—ã—Ö –≤–µ—Å–æ–≤ *w_init*, —Å–æ—Å—Ç–æ—è—â–∏–π –∏–∑ –Ω—É–ª–µ–π. –û—Å—Ç–∞–≤—å—Ç–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã  *eta* –∏ *seed* —Ä–∞–≤–Ω—ã–º–∏ –∏—Ö –∑–Ω–∞—á–µ–Ω–∏—è–º –ø–æ —É–º–æ–ª—á–∞–Ω–∏—é (*eta*=0.01, *seed*=42 - —ç—Ç–æ –≤–∞–∂–Ω–æ –¥–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏ –æ—Ç–≤–µ—Ç–æ–≤).**

```python
stoch_grad_desc_weights, stoch_errors_by_iter = stochastic_gradient_descent(X2, y, [0,0,0,0], max_iter=1e5)
```
