```python
import numpy as np
from matplotlib import pyplot as plt
from scipy import optimize
```

```python
def func(x):
    return np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2) 
```

```python
x = np.array(np.arange(1, 31, 1))
```

```python
plt.plot(x, func(x))
```

```python
a1 = round(optimize.minimize(func, 2, method = 'BFGS').fun, 2)
a2 = round(optimize.minimize(func, 30, method = 'BFGS').fun, 2)
```

```python
with open('answer_3_1.txt', 'w') as answer1:
    answer1.write(f'{a1} {a2}')
```

```python
a3 = round(optimize.differential_evolution(func, [(1, 30)]).fun[0], 2)
```

```python
with open('answer_3_2.txt', 'w') as answer2:
    answer2.write(f'{a3}')
```

```python
def func2(x):
    return func(x).astype(int)
```

```python
plt.plot(x, func(x))
plt.plot(x, func2(x))
plt.show()
```

```python
a4 = round(optimize.minimize(func2, 30, method = 'BFGS').fun[0], 2)
a5 = round(optimize.differential_evolution(func2, [(1, 30)]).fun, 2)
```

```python
with open('answer_3_3.txt', 'w') as answer3:
    answer3.write(f'{a4} {a5}')
```

```python
print(f'{a1} {a2} {a3} {a4} {a5}')
```
