```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

```python
k = 7
```

```python
#arr = np.random.chisquare(1, 1000)
chi_rv = sts.chi2(k)
chi_table = chi_rv.rvs(size=1000)
```

```python
plt.hist(chi_table, bins = 25, density = True)

x = np.linspace(0,25,100)
chi_teor = sts.chi2(k)
pdf = chi_teor.pdf(x)
plt.plot(x, pdf)
```

```python
med = np.array([])
for i in range(0,1000):
    chi = sts.chi2(k)
    chi_t = chi.rvs(5)
    medium = sum(chi_t)/len(chi_t)
    med = np.append(med, medium)
plt.hist(med, bins = 30, density = True)
```
